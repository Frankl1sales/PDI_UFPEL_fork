```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.color import rgb2gray
```

```python
# Carregue sua imagem
cargoship_rgb = imread('images/manila-bay.jpg')

# Converta para escala de cinza
cargoship_gray = rgb2gray(cargoship_rgb)

# Visualize
plt.imshow(cargoship_gray, cmap='gray')
plt.title('Cargo Ship in Grayscale')
plt.axis('off')
plt.show()
```

```python
cargoship_fft = np.fft.fftshift(np.fft.fft2(cargoship_gray))

# Visualize o espectro
plt.imshow(np.log(abs(cargoship_fft)), cmap='gray')
plt.title('Fourier Transform Representation')
plt.axis('off')
plt.show()
```

```python
# Máscara vertical (ajuste os índices conforme necessário)
cargoship_fft[:1335, 1820:1828] = 1
cargoship_fft[-1335:, 1820:1828] = 1

# Visualização da FFT modificada
plt.imshow(np.log(abs(cargoship_fft)), cmap='gray')
plt.title('Modified Fourier Transform')
plt.axis('off')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 7))

ax[0].imshow(cargoship_gray, cmap='gray')
ax[0].set_title('Original Grayscale Image')
ax[0].axis('off')

transformed_image = abs(np.fft.ifft2(np.fft.ifftshift(cargoship_fft)))
ax[1].imshow(transformed_image, cmap='gray')
ax[1].set_title('Transformed Image')
ax[1].axis('off')

plt.show()
```

```python
def rgb_transform(img, channel):
    fft = np.fft.fftshift(np.fft.fft2(img[:, :, channel]))
    fft[:1335, 1820:1828] = 1
    fft[-1335:, 1820:1828] = 1
    return abs(np.fft.ifft2(np.fft.ifftshift(fft))).clip(0, 255)

# Aplicar em cada canal
red = rgb_transform(cargoship_rgb, 0)
green = rgb_transform(cargoship_rgb, 1)
blue = rgb_transform(cargoship_rgb, 2)

# Reconstrução
new_image = np.dstack([red, green, blue]).astype(np.uint8)

# Visualização
fig, ax = plt.subplots(1, 2, figsize=(15, 7))
ax[0].imshow(cargoship_rgb)
ax[0].set_title('Original RGB')
ax[0].axis('off')

ax[1].imshow(new_image)
ax[1].set_title('Transformed RGB')
ax[1].axis('off')

plt.show()
```
