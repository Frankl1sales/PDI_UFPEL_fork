```python
import matplotlib.pyplot as plt
from skimage import io, img_as_float
from scipy.ndimage import convolve
import numpy as np

# 1. Ler a imagem
I = io.imread('images/cameraman.tif')

# 2. Converter a imagem para float (valores entre 0 e 1)
I_float = img_as_float(I)

# 3. Definir o filtro passa-baixa (máscara de média 3x3)
kernel = np.ones((3, 3)) / 9.0

# 4. Aplicar o filtro
I_lowpass = convolve(I_float, kernel)

# 5. Exibir as imagens
plt.figure(figsize=(10, 8))

# Imagem original
plt.subplot(1, 2, 1)
plt.imshow(I, cmap='gray', vmin=0, vmax=255)
plt.title('Imagem Original')
plt.axis('off')

# Imagem filtrada (passa-baixa)
plt.subplot(1, 2, 2)
plt.imshow(I_lowpass, cmap='gray')
plt.title('Imagem com Filtro Passa-Baixa')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from skimage import io, img_as_float
from scipy.ndimage import convolve
import numpy as np

# 1. Ler a imagem
I = io.imread('images/cameraman.tif')

# 2. Converter para float (valores entre 0 e 1)
I_float = img_as_float(I)

# 3. Definir o kernel de passa-alta (realce)
kernel_highpass = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

# 4. Aplicar o filtro
I_highpass = convolve(I_float, kernel_highpass)

I_highpass = np.clip(I_highpass, 0, 1)

# 5. Exibir resultados
plt.figure(figsize=(10, 5))

# Imagem original
plt.subplot(1, 2, 1)
plt.imshow(I, cmap='gray', vmin=0, vmax=255)
plt.title('Imagem Original')
plt.axis('off')

# Imagem com filtro passa-alta
plt.subplot(1, 2, 2)
plt.imshow(I_highpass, cmap='gray')
plt.title('Filtro Passa-Alta')
plt.axis('off')

plt.tight_layout()
plt.show()
```
